```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\sathv\Downloads\Summer-Olympic-medals-1976-to-2008.csv", encoding='latin1')
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df= df.drop(columns=['Country_code', 'Event_gender'],errors='ignore')
```

```python
df= df.apply(lambda x: x.str.strip() if x.dtype== "object" else x)
```

```python
df= df.drop_duplicates()
```

```python
df.head()
df.info()
```

```python
#Top 10 countries by Medals
top_countries= df['Country'].value_counts().head(10)
print(top_countries)

top_countries.plot(kind="bar", figsize=(8,4), title="Top 10 Countries by Medal Count")
plt.ylabel("Medal Count")
plt.show()
plt.savefig('top_countries.png', bbox_inches='tight')    
plt.close() 
```

```python
#Top 10 Athletes with Most Medals

top_athletes = df["Athlete"].value_counts().head(10)
print(top_athletes)

top_athletes.plot.bar(figsize=(8,4), title= "Top 10 Athletes by Medal Count")
plt.xlabel("Athlete")
plt.ylabel("Medal Count")
plt.savefig("top_athletes.png", bbox_inches="tight")
plt.show()
```

```python
#Gender Distribution in Event 

gender_count= df["Gender"].value_counts()
print(gender_count)
gender_count.plot(kind="pie",color=["pink","lightblue"],autopct="%1.1f%%",startangle=90, figsize=(8,4))
plt.title("Medal by Gender")
plt.savefig("gender_count.png")
plt.show()
```

```python
#Medals by Sport

top_sports= df["Sport"].value_counts().head(15)
print(top_sports)

top_sports.plot.barh(color="green", figsize=(8,4))
plt.title("Top 15 Sports Medals")
plt.xlabel("Medal Count")
plt.savefig("top_sports.png")
plt.show()
```

```python
#Total Medals by each country per year
country_year= df.groupby(["Country","Year"])["Medal"].count().reset_index()
country_year.head()
```

```python
#Creating a pivot table 
pivot_table= pd.pivot_table(df,
                            index="Country",
                            columns="Year",
                            values="Medal",
                            aggfunc="count",
                            fill_value=0)
pivot_table.head()
                        
                            
```

```python
#Visualize the trend for a top single country
usa_trend= country_year[country_year["Country"]=="United States"]

plt.figure(figsize=(8,4))
plt.plot(usa_trend["Year"], usa_trend["Medal"], marker="o", color="yellow")
plt.title("USA Medals Over the year")
plt.xlabel("Year")
plt.ylabel("Medal Count")
plt.savefig("usa_trend.png")
plt.show()
```

```python
#Top Countries per Year
top_country_per_year=(
    country_year.sort_values(["Year","Medal"],ascending=[True,False])
    .groupby("Year").first().reset_index()
)
top_country_per_year.head(10)
```

```python
#Medals by Year

medals_by_year = df['Year'].value_counts().sort_index()
plt.figure(figsize=(8,4))
plt.plot(medals_by_year.index, medals_by_year.values, marker='o')
plt.title('Total Medals by Year')
plt.xlabel('Year')
plt.ylabel('Number of Medals')
plt.savefig("medals_by_year.png")
plt.show()
```
